```python
import json
from pathlib import Path
result = {'datasets_top_level': [], 'normal_trace_csv': [], 'failure_trace_csv': []}
datasets_dir = Path('/kaggle/input/datasets')
if datasets_dir.exists():
    try:
        result['datasets_top_level'] = sorted(p.name for p in datasets_dir.iterdir())
        # Walk 3 levels deep
        for path in datasets_dir.rglob('*'):
            if path.is_file():
                if path.name == 'normal_trace.csv':
                    result['normal_trace_csv'].append(str(path))
                if path.name == 'failure_trace.csv':
                    result['failure_trace_csv'].append(str(path))
    except Exception as e:
        result['datasets_iterdir_error'] = str(e)
out = Path('/kaggle/working/kaggle_diag2.json')
out.write_text(json.dumps(result, indent=2, default=str))
print('WROTE:', out)
print(json.dumps(result, indent=2, default=str)[:3000])
```
